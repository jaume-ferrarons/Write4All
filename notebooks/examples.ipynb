# Examples

In this notebook I show how the suggestions to improve some examples.

```python
# Add parent directory to path
import os
import sys
sys.path.append(os.path.abspath('..'))

from IPython.display import display, HTML, Markdown
import PIL
from review import process_text, process_image

gemini_1_0 = "Gemini 1.0 Pro"
gemini_1_5 = "Gemini 1.5 Pro (latest)"
gemini_1_0_vision = "Gemini 1.0 Pro Vision"
```

## Text examples

```python
def demo_text(original_text: str):
    display(HTML(f"<h1>Original text</h1>{original_text}"))
    display(HTML(process_text(text=original_text, model=gemini_1_5)))

text_examples = [
    "Mankind has always been fascinated by the stars.",
    "The blacklist is too large.",
    "The average user prefers his information presented visually.",
    "Native English speaker.",
    "I'm totally OCD about keeping my desk clean.",
    "People of that religion are always so conservative.",
    "You speak English really well for a foreigner.",
    "He's too skinny to play sports.",
    "This sentence is in English.",
]
for text in text_examples:
    demo_text(text)
```

## Images

```python
def demo_image(image_path: str):
    # display(Image(image_path, width=600))
    image = PIL.Image.open(image_path)
    image.thumbnail((600, 600))
    display(image)
    response = process_image(image=image, model=gemini_1_0_vision)
    display(Markdown(response))

demo_image("../static/images/CEOs.png")
demo_image("../static/images/meat_grid.png")
demo_image("../static/images/crosses.jpg") #Â Author: Gerd Altmann - https://pixabay.com/users/geralt-9301/
demo_image("../static/images/elephants.jpg") #Â Author: Gerd Altmann - https://pixabay.com/users/geralt-9301/
```
